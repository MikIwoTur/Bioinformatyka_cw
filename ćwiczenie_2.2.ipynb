```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

  $ \left ( {0,5 \cdot x^2 - 2,5 \cdot x + 0,5} \right ) + \left ( 8,5 \cdot \sin x \right ) $

```python
x = np.arange(-10, 15, 0.2)
y_square = 0.5 * x*x - 2.5 * x + 0.5
y_sin = 8.5 * np.sin(x)
y = y_sin + y_square
fig = plt.figure(figsize=(5, 4), dpi=150)
ax = fig.add_axes([0.15, 0.15, 0.8, 0.8])
ax.plot(x, y_square, color='red', linewidth=0.8, alpha=0.5, label='Square')
ax.plot(x, y_sin, color='blue', linewidth=0.8, alpha=0.5, label='Sin')
ax.plot(x, y, marker='o', markersize=3, markerfacecolor='red', markeredgecolor='blue', markeredgewidth=0.5, linestyle=' ', label='Sum')
ax.legend()
ax.set_xlabel('oś X')
ax.set_ylabel('oś Y')
ax.set_title("(0,5⋅x^2−2,5⋅x+0,5)+(8,5⋅sinx)")
ax.savefig('output/chart_2.png', transparent=False, dpi=300)
```

```python
x = np.arange(-10, 15, 0.2)
y_square = 0.5 * x*x - 2.5 * x + 0.5
y_sin = 8.5 * np.sin(x)
y = y_sin + y_square
fig = plt.figure(figsize=(5, 4), dpi=150)
ax = fig.add_axes([0.15, 0.15, 0.8, 0.8])
ax.plot(x, y_square, color='red', linewidth=0.8, alpha=0.5, label='Square')
ax.plot(x, y_sin, color='blue', linewidth=0.8, alpha=0.5, label='Sin')
ax.plot(x, y, marker='o', markersize=3, markerfacecolor='red',
        markeredgecolor='blue', markeredgewidth=0.5, linestyle=' ', label='Sum')
ax.legend()
ax.set_xlabel('oś X')
ax.set_ylabel('oś Y')
ax.set_title("(0,5⋅x^2-2,5⋅x+0,5)+(8,5⋅sinx)")
ax.set_xlim(right=43)
ax.set_xticks(np.arange(0, 100))
ax.set_ylim(top=250)
ax.set_yticks(np.arange(-50, 200))
plt.show()
```
